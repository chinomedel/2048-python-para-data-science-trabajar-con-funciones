## **¿Cuáles son los valores promedio de alquiler por tipo de propiedad**

```python
datos.head()
```

```python
datos['Valor'].mean() ## Es el promedio de todos los inmuebles NO por inmueble
```

```python
datos.groupby('Tipo').mean(numeric_only=True) ## Trae el promedio por inmueble pero de todas las columnas y solo queremos el de la columna Valor
```

```python
datos.groupby('Tipo')['Valor'].mean(numeric_only=True)
```

```python
df_tipo_precio = datos.groupby('Tipo')[['Valor']].mean(numeric_only=True).sort_values('Valor')
```

```python
df_tipo_precio.plot(kind='barh', figsize=(12,5), color='green')
```

## **¿Cuál es el porcentaje de cada tipo de propiedad en nuestra base de datos?"**

```python
datos.Tipo.unique()
#df_tipo_precio.query('@inmuebles_comerciales in Tipo')
```

```python
comerciales = ['Conjunto Comercial/Sala','Edificio Completo', 'Departamento en Hotel',
                'Tienda/Salón', 'Casa Comercial','Terreno Estándar', 'Cochera/Estacionamiento',
               'Galpón/Depósito/Almacén', 'Tienda en Centro Comercial','Posada/Chalé', 'Estudio', 'Hotel','Industria']
```

```python
casas_residenciales = datos.query('@comerciales not in Tipo')
```

```python
promedio_precio_casas_residenciales = casas_residenciales.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
promedio_precio_casas_residenciales.plot(kind='barh')
```

```python
from os import name
porcentaje_casas_residenciales = casas_residenciales.Tipo.value_counts('Normalize')
porcentaje_casas_residenciales.plot(kind='barh', figsize=(12,8), color='green', xlabel="Tipo Inmueble", ylabel="Porcentaje")
```

```python
solo_departamentos = casas_residenciales.query('Tipo == "Departamento"')
solo_departamentos
```

```python
solo_departamentos['Habitaciones'].mean()
```

```python
solo_departamentos.Colonia.nunique()
```

```python
precio_por_colonia = solo_departamentos.groupby('Colonia')[['Valor']].mean(numeric_only=True).sort_values('Valor').head(10)
precio_por_colonia.plot(kind="bar")
```
